```python
import pandas as pd
```

```python
t = pd.read_csv("산사태_현황.csv")
```

```python
t
```

```python
t=t.loc[t["산사태재해년도"]==2020]
```

```python
col = ["시도명","읍면명","산사태재해명","산사태재해년도"]
```

```python
t.drop(col,axis=1,inplace=True)
```

```python
t.reset_index(drop=True)
```

```python
tp = t["산사태피해면적"].groupby(t["시군구명"])
```

```python
tp=tp.sum().to_frame()
```

```python
tp
```

```python
tp=tp.pivot_table(index=['시군구명',"산사태피해면적"])
```

```python
tp.reset_index(inplace=True)
```

```python
tp.rename(columns={"산사태피해면적":"2020년_산사태피해면적","시군구명":"시군구"},inplace=True)
```

```python
t1=t["시군구명"].value_counts().to_frame()
```

```python
t1.rename(columns={"시군구명":'2020년_산사태_발생건수'},inplace=True)
```

```python
t1=t1.reset_index().rename(columns={"index":"시군구"})
```

```python
ttp=pd.merge(t1,tp, how='outer',on="시군구")
```

```python
ttp
```

```python
ttp.to_csv("산사태_전처리.csv", index=False, encoding="utf-8-sig")
```
