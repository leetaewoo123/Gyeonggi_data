```python
import pandas as pd
```

```python
t = pd.read_csv("2022_터널현황_원본.csv")
```

```python
# row 생략 없이 출력
pd.set_option('display.max_rows', None)
# col 생략 없이 출력
pd.set_option('display.max_columns', None)
```

```python
t.head(5)
```

```python
t.shape
```

```python
t = t.loc[t["시도"]=="경기도"]
```

```python
t.isna().sum()
```

```python
col_t = ["노선명","리",'유효폭','교통량','준공년도','높이','총폭','도로종류','시설명',"읍면동"]
```

```python
t.drop(col_t,axis=1,inplace=True)
```

```python
t.reset_index(drop=True,inplace=True)
```

```python
tg=t['총길이'].groupby(t['시군구'])
```

```python
tg=tg.sum().to_frame()
```

```python
tg=tg.pivot_table(index=['시군구','총길이'])
```

```python
tg.reset_index(inplace=True)
```

```python
tg
```

```python
t.reset_index(drop=True,inplace=True)
```

```python
t1=t["시군구"].value_counts().to_frame()
```

```python
t1.rename(columns={"시군구":'터널수'},inplace=True)
```

```python
t1
```

```python
t1=t1.reset_index().rename(columns={"index":"시군구"})
```

```python
ttg=pd.merge(t1,tg, how='outer',on="시군구")
```

```python
ttg
```

```python
ttg.to_csv("터널_전처리.csv", index=False, encoding="utf-8-sig")
```
