```python
# 사용자 운영체제 확인
import os
os.name

# 운영체제별 한글 폰트 설정
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')

plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

# 글씨 선명하게 출력하기 
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('(총정리)경기도산사태위험지역 데이터.csv', encoding ='utf-8',index_col=0)
df.head(3)
```

```python
df[['2019년 평균 강수량','2020년 평균 강수량']]
```

```python


plt.figure(figsize=(10, 5))
sns.lineplot( y="2019년 평균 강수량", x="시군구",data=df,marker='o',
         markersize=6 )
sns.lineplot( y="2020년 평균 강수량", x="시군구",data=df,marker='o', 
         markersize=6)

plt.title('2019/2020 평균 강수량 비교', fontsize=15)

plt.xlabel('시군구',fontsize=1);
plt.ylabel('평균강수량',fontsize=12);

plt.show()
```
