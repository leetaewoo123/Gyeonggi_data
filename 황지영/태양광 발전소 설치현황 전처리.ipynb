# 불필요한 columns 제거

```python
df=df.drop(['세부용도','설치위치구분','허가기관명','용도','설비용량(KW)','주파수(Hz)','발전소명','사업개시예정일자','사업개시일자','집광판면적(㎡)','비고','데이터기준일자','위도','경도'],axis=1)#전체 열 삭제
```

```python
list=['공사계획신고','설치불가','자진반납','합병말소','허가취소','폐업']

for i in list:
    A = df[df['사업상태'] == i].index
    df.drop(A,axis='index',inplace=True)
```

```python
df['사업상태'].value_counts()
```

```python
list2=['미운영','인허가취소','인허가']

#폐업,미운영,인허가취소','인허가'=>운영중으로 바꾸기
for i in list2:
    df.loc[df['사업상태']==i,'사업상태'].fillna('운영중')
    B = df[df['사업상태'] == i].index
    df.drop(B,axis='index',inplace=True)
```

```python
df['사업상태'].value_counts()
```

```python
df['지목']=df['지목'].fillna('no')# 넣어주지 않으면 오류발생
```

```python
df['소재지지번주소']=df['소재지지번주소'].fillna('None')
df['소재지도로명주소']=df['소재지도로명주소'].fillna('None')
df
```

```python
# 포함하고자 하는 문자열 리스트 생성
list = ['임야','구거','도로','하천','철도','임','no']
# 리스트 안의 문자가 포함되면 열 추가
new_df=pd.DataFrame()
for i in list:
    df1=df.loc[df['지목'].str.contains(i)]
    new_df=new_df.append(df1,ignore_index=True)
new_df
```

```python
new_df['지목'].value_counts()
```

```python
new_df
```

```python
#지목이 'none'인 값이 너무 많기 때문에  none중' 산 '이 포함된 행은 임야로 바꿔줌 
mountain=new_df.loc[new_df['소재지지번주소'].str.contains(' 산 ',na=False)|new_df['소재지도로명주소'].str.contains(' 산 ',na=False)]
```

```python
mountain['지목']=mountain['지목'].str.replace('no','임야')
```

```python
#원래 데이터에서 산을 더함
df=pd.concat([new_df,mountain],axis=0)
df
```

```python
df=df.drop_duplicates()
```

```python
df=df.loc[ (df['지목'] !='no') ]
df.reset_index(drop=True)
```
