```python
import webbrowser
import folium, json
import pandas as pd

m = folium.Map(location=[36, 127], tiles="OpenStreetMap", zoom_start=8)
geo_data = json.load(open('sido.geojson', encoding='utf-8'))

df = pd.read_csv('sad.csv', encoding='utf-8', dtype={'code':'str'})
folium.Choropleth(geo_data=geo_data,
                  data = df,
                  columns=['sigun_nm', 'pay'],
                  key_on='feature.properties.name',
                  fill_color='RdPu',
                  fill_opacity=0.7,
                  line_opacity=0.5
                  ).add_to(m)

m.save('folium_kr.html')
webbrowser.open_new("folium_kr.html")
```
