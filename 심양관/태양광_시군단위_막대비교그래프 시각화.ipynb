```python
t.reset_index(drop=True,inplace=True)
```

```python
t["지역명"]= t['시군명']
```

```python
t1=t["지역명"].value_counts().to_frame()
```

```python
t1.rename(columns={"지역명":'태양광발전소 수'},inplace=True)
```

```python
t1=t1.reset_index().rename(columns={"index":"지역명"})
```

```python
p["지역명"]= p['시군구명']
```

```python
p1 = p['지역명'].value_counts().to_frame()
```

```python
p1.rename(columns={"지역명":'산사태발생횟수'},inplace=True)
```

```python
p1=p1.reset_index().rename(columns={"index":"지역명"})
```

```python
tp = pd.merge(t1,p1,how='outer',on='지역명')
```

```python
tp.fillna(0,inplace=True)
```
